```python
fig, ax = plt.subplots(figsize=(11, 9))

# plot heatmap
sb.heatmap(df, cmap= sb.diverging_palette(0, 230, 90, 60, as_cmap=True), vmin= -1, vmax=1,
           annot=True, annot_kws={"size":15}, fmt=".2f", linewidth=0.3, cbar_kws={"shrink": .8})

ax.xaxis.tick_top()
xticks_labels = ['Postchild Elementary', 'Skilled', 'Sales', 'Serviece', 'Clerk', 'Professional', 'Manager']
plt.xticks(np.arange(7) + .5, labels=xticks_labels)

# axis labels
plt.xlabel('After childbirth')
plt.ylabel('Before childbirth')

plt.show()
```
